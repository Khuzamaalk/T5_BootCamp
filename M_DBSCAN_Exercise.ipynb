#**DBSCAN**


#Step 1 : Import Libraries



```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn import datasets
```

#Step 2 : Prepare dataset
###**use make blobs with different numbers**



```python
iris = load_iris()

# Split the data into features and target variable
iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
'''X = iris.data
y = iris.target'''
```

#Step 3 : Modeling The Data Using DBSCAN



```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5) ##كلما قللنا الايبسلون كلما زادت الاوتلايرز

dbscan.fit(df[iris.feature_names])

labels = dbscan.labels_

df['labels'] = labels
```

#Step 4 : Evaluation Metrics For DBSCAN Algorithm In Machine Learning


```python
from sklearn.metrics import silhouette_score
print(silhouette_score(df[iris.feature_names], labels))


from sklearn.metrics import adjusted_rand_score
print(adjusted_rand_score(df['target'], labels))

import matplotlib.pyplot as plt
plt.scatter(df[iris.feature_names[0]], df[iris.feature_names[1]], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
