#**Python Code Implementation of t-SNE on MNIST Dataset**


#Step 1 :
##Importing Necessary Modules.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

# Step 2 :
## load the MNIST_test dataset into pandas dataframe

```python
df = pd.read_csv('/content/mnist_train.csv')

df.head()
```

```python
X = df.drop("label", axis = 1)
y = df['label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state = 42)
```

# Step 3  :
## Standardize the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train.shape)
```

# Step 4 :  Reduce the columns data to 2 dimensions




```python
X_train_1000 = X_train[0:1000, :]
labels = y.values
labels_1000 = labels[0:1000]
```

# Step 5 :
## define the model

```python
model = TSNE(n_components= 2, random_state= 0)
```

# Step 6 :
## Fit the model

```python
tsne_data = model.fit_transform(X_train_1000)
tsne_data = np.vstack((tsne_data.T, labels_1000)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "label"))
```

# Step 7 :
##  Plotting the result

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=tsne_df, x='Dim_1', y='Dim_2', hue='label', palette='bright')
plt.show()
```
